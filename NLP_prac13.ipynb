```python
import torch
from transformers import GPT2Tokenizer
from transformers import GPT2LMHeadModel

# Initialize the tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("microsoft/DialoGPT-medium")

model = GPT2LMHeadModel.from_pretrained("microsoft/DialoGPT-medium")

# Generate a response given an input message
def generate_response(user_input, chat_history_ids):
  # Encode the user input and add the eos_token
  user_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors="pt")

  # Reshape user_input_ids to match the dimensions of chat_history_ids
  user_input_ids = user_input_ids.squeeze()

  # Convert chat_history_ids to a tensor
  chat_history_ids_tensor = torch. tensor(chat_history_ids, dtype=torch. long)

  # Append the user input tokens to the chat history
  bot_input_ids = torch.cat([chat_history_ids_tensor, user_input_ids], dim =- 1)

  # Generate a response while limiting the total response length
  output = model.generate(bot_input_ids.unsqueeze(0), max_length=1000, pad_token_id=tokenizer.eos_token_id)

  # Decode the generated response and convert it to a string
  response = tokenizer.decode(output[0, bot_input_ids. shape[-1]: ], skip_special_tokens=True)
  return response
```

```python
print("Chatbot: Hi! How can I assist you today?")
chat_history_ids = []

while True:
  user_input = input("User: ")

  # Generate a response
  response = generate_response(user_input, chat_history_ids)

  # Print the response
  print("Chatbot:", response)

  # Update the chat history
  chat_history_ids.extend(tokenizer.encode(response + tokenizer.eos_token))
```

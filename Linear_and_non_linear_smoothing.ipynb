```python
from PIL import Image,ImageFilter
from skimage.color import rgb2gray
from skimage.io import imread,imshow,show
import matplotlib.pylab as pylab
```

```python
#Non linear smoothing

#Min Filter

im = Image.open('./images/parrot.jpg')

im1 = im.filter(ImageFilter.MinFilter())

pylab.imshow(im1)
pylab.show()
```

```python
#Max filter
im = Image.open('./images/parrot.jpg')

im1 = im.filter(ImageFilter.MaxFilter())
pylab.imshow(im1)
pylab.show()
```

```python
#Median Filter
im = Image.open('./images/parrot.jpg')

im1 = im.filter(ImageFilter.MedianFilter(5))

pylab.imshow(im1)
pylab.show()
```

```python
#Linear Smoothing

#Gaussian Blur Filter

im = Image.open('./images/parrot.jpg')

im1 = im.filter(ImageFilter.GaussianBlur())

pylab.imshow(im1)
pylab.show()
```

```python
#Blur Filter
im = Image.open('./images/parrot.jpg')

im1 = im.filter(ImageFilter.BLUR())

pylab.imshow(im1)
pylab.show()
```

```python
from PIL import Image 
from skimage.io import imread,imshow,show
import matplotlib.pylab as pylab
import numpy as np
from scipy import misc,signal
from skimage.color import rgb2gray
```

```python
face_image = misc.face(gray=True) - misc.face(gray=True).mean()
template_image = np.copy(face_image[300:365,600:750])
template_image -= template_image.mean()

correlated_image = signal.correlate2d(face_image,template_image)

pylab.imshow(correlated_image,cmap='afmhot')
pylab.show()
```

```python
im = rgb2gray(imread('./images/parrot.jpg'))

im = im.astype(float)

edge_laplase_kernel = [[0,1,0],[1,-4,1],[0,1,0]]

blur_box_kernel =np.ones((3,3))/9

im1 = np.clip(signal.convolve2d(im,edge_laplase_kernel),0,1) 

im2 = signal.convolve2d(im,blur_box_kernel)

pylab.imshow(im2)
pylab.show()
```

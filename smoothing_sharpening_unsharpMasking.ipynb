```python
from PIL import Image
from skimage.io import imread,imshow,show
from scipy import misc
from scipy import ndimage
import numpy as np
from skimage.color import rgb2gray
from skimage import img_as_float
import matplotlib.pylab as pylab
```

```python
#unsharp masking

im = rgb2gray(img_as_float(imread('./images/parrot.jpg')))
im1 = ndimage.gaussian_filter(im,3)
im2 = np.clip(im-im1,0,1)
pylab.imshow(im2)
pylab.show()
```

```python
#sharpening

from skimage.filters import laplace

im = rgb2gray(imread('./images/parrot.jpg'))
im1 = np.clip(laplace(im)+im,0,1)
pylab.imshow(im1)
pylab.show()
```

```python
#smoothing

im = rgb2gray(imread('./images/parrot.jpg'))
k=5
im1 = ndimage.percentile_filter(im,percentile=5,size=5)
pylab.imshow(im)
pylab.show()
```
